```python
from sympy import *
from sympy.plotting import plot
```

```python
init_printing(use_unicode=False, wrap_line=False, no_global=True)
```

```python
x = Symbol('x')
f = x + 1
f
```

```python
f.subs(x, 0)
```

```python
f.subs(x, -1)
```

```python
plot(f)
```

```python
f = -3 * x + 5
```

```python
f.subs(x, 0)
```

```python
f.subs(x, 1)
```

```python
plot(f)
```
